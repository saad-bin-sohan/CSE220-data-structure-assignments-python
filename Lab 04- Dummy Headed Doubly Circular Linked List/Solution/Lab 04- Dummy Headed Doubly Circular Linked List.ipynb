```python
class Node:
  def __init__(self, e, n, p):
    self.element = e
    self.next = n
    self.prev = p
```

```python
from IPython.core.history import HistorySavingThread
class DoublyList:

  # Creates a Dummy Head for Dummy Headed Circular Doubly Circular Linked List.
  def __init__(self):

    dummy_head=Node(None,None,None)
    self.dummy_head=dummy_head
    dummy_head.next=dummy_head
    dummy_head.prev=dummy_head
    self.tail=dummy_head



  #create a dummy_headDCL from an array
  def create_list(self, l):

    for i in range(len(l)):
      n=Node(l[i],self.dummy_head,self.tail)
      self.tail.next=n
      self.tail=self.tail.next
      self.dummy_head.prev=self.tail


  # Counts the number of Nodes in the list and return the number
  def countNode(self):

    c=0
    temp=self.dummy_head.next
    while temp!=self.dummy_head:
      c+=1
      temp=temp.next

    return c



  # prints the elements in the list forward
  def forwardprint(self):

    temp=self.dummy_head.next
    while temp!=self.dummy_head:
      if temp.next==self.dummy_head:
        print(temp.element)
        break
      print(temp.element,end=",")
      temp=temp.next



  # prints the elements in the list backward
  def backwardprint(self):

    temp=self.dummy_head.prev
    while temp!=self.dummy_head:
      if temp.prev==self.dummy_head:
        print(temp.element)
        break
      print(temp.element,end=",")
      temp=temp.prev



  # returns the reference of the at the given index. For invalid index return None.
  def nodeAt(self, idx):

    c=0
    temp=self.dummy_head.next
    while temp!=self.dummy_head:
      if c==idx:
        t=True
        return temp
      temp=temp.next
      c+=1
    return None




  # returns the index of the containing the given element. if the element does not exist in the List, return -1.
  def indexOf(self, elem):

    temp=self.dummy_head.next
    c=0
    while temp!=self.dummy_head:
      if temp.element==elem:
        return c
      temp=temp.next
      c+=1
    return -1



  # insert at specific index of the dummy_headDCL
  def insert(self, elem, idx):

    newnode=Node(elem,None,None)
    c=0
    temp=self.dummy_head.next
    ptemp=self.dummy_head
    while temp!=self.dummy_head:
      if idx==c:
        ptemp.next=newnode
        newnode.prev=ptemp
        newnode.next=temp
        temp.prev=newnode
        break
      temp=temp.next
      ptemp=ptemp.next
      c+=1

  #removes the node of the specific index and returns the element of the node
  def remove(self, idx):

    c=0
    temp=self.dummy_head.next
    ptemp=self.dummy_head
    while temp!=self.dummy_head:
      if idx==c:
        ptemp.next=temp.next
        ptemp.next.prev=ptemp
        return temp.element
      temp=temp.next
      ptemp=ptemp.next
      c+=1
```

```python
class Patient:
  def __init__(self,id=None,name=None,age=None,bg=None):
    self.id=id
    self.name=name
    self.age=age
    self.bg=bg


class WRM:
  def __init__(self):
    self.dummy_head=Patient()
    self.dummy_head.next=self.dummy_head
    self.dummy_head.prev=self.dummy_head
    self.tail=self.dummy_head


  def RegisterPatient(self,id,name,age,bg):
    obj = Patient(id, name, age, bg)
    obj.next = self.dummy_head
    obj.prev = self.tail
    self.tail.next = obj
    self.tail = obj


  def ServePatient(self):
    print(self.dummy_head.next.name,"just visited the doctor")
    self.dummy_head.next=self.dummy_head.next.next



  def CancelAll(self):
    self.dummy_head.next=None
    self.dummy_head.prev=None
    print("Success, all the appointments have been canceled")

  def CanDoctorGoHome(self):
    if self.dummy_head.next==None:
      return "Yes, doctor can go home."
    return "No, there are more patients left"

  def ShowAllPatient(self):
    temp=self.dummy_head.next
    if self.dummy_head.next==None:
      print("No one to show")
    else:
      while temp!=self.dummy_head:
        print("List of all patients:")
        print("ID:",temp.id)
        print("Name:",temp.name)
        print("Age:",temp.age)
        print("Blood Group:",temp.bg)
        temp=temp.next

#Tester
waiting=WRM()
print("Choose one of the below:")
print("1)Register a ptient")
print("2)Serve a patient")
print("3)Cancel all appointment")
print("4)Can doctor go home")
print("5)Show all patients information")
print("6)Exit")


while True:
  inp= int(input("Which number have you picked? "))
  if inp==6:
    break
  elif inp==1:
    a,b,c,d=int(input("ID:")),input("Name:"),int(input("Age:")),input("Blood Group:")
    waiting.RegisterPatient(a,b,c,d)
  elif inp==2:
    waiting.ServePatient()
  elif inp==3:
    waiting.CancelAll()
  elif inp==4:
    print(waiting.CanDoctorGoHome())
  elif inp==5:
    waiting.ShowAllPatient()
```
