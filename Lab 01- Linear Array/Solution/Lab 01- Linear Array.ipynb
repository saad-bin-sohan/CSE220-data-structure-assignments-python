```python
# Complete the functions defined in this cell


# Test 01: Play Right
def playRight(sequence,beats):
  # TO DO

  def swapRight():
    temp = sequence[len(sequence)-1]
    for i in range(-1, -len(sequence), -1):
      sequence[i] = sequence[i-1]
    sequence[0] = temp
    return sequence

  for elem in beats:
    if elem == 1:
      swapRight()
  return sequence








# Test 02: Discard Cards
def discardCards(cards,number):
  # TO DO

  def helper(idx):
    cards[idx] = 0
    return cards

  for i in range(len(cards)):
    if cards[i] == number:
      helper(i)

  for i in range(len(cards)):
    for j in range(i+1, len(cards)):
      if cards[j]!= 0 and cards[i] == 0:
        cards[i], cards[j] = cards[j], cards[i]
        break

  return cards








#Test 03: Merge Lineup
def mergeLineup(pokemon_1, pokemon_2):
  # TO DO
  combined_array = np.array([0]*len(pokemon_1))

  for i in range(len(pokemon_1)):

    if pokemon_1[i] == None:
      num_1 = 0
    else:
      num_1 = pokemon_1[i]

    if pokemon_2[len(pokemon_2) - i - 1] == None:
      num_2 = 0

    else:
      num_2 = pokemon_2[ len(pokemon_2) - i - 1]

    combined_array[i] = num_1 + num_2

  return combined_array








# Test 04: Balance your Salami
def balanceSalami(salami):
  # TO DO
  def totalSalamiCounter(arr):
    sum = 0
    for taka in arr:
      sum += taka
    return sum

  sum = totalSalamiCounter(salami)

  def divisibility_meseaurer(salami, sum):
    right_person_gets = sum
    left_person_gets = 0

    for elem in salami:
      right_person_gets -= elem
      left_person_gets += elem

      if right_person_gets == left_person_gets:
        return True

    return False

  result = divisibility_meseaurer(salami, sum)

  return result







# Test 05: Protecc Salami

def protectSalami(salami):
  # TO DO
  unique_elements = np.array([0]*len(salami))
  frequency_holder = np.array([0]*len(salami))

  count_1 = 0
  for elem in salami:
    if elem not in unique_elements:
      unique_elements[count_1] = elem
      count_1 += 1

  count_2 = 0
  for elem in unique_elements:
    for elm in salami:
      if elem == elm:
        frequency_holder[count_2] += 1
    count_2 += 1

  result = False

  for i in range(len(frequency_holder)):
    for j in range(i+1, len(frequency_holder)):

      if (frequency_holder[i] == frequency_holder[j]) and (frequency_holder[i] != 0) and (frequency_holder[i] != 1) and (unique_elements[i] != 0):
        result = True

        return result

  return result








# Test 06: Odd Even Wave
def waveYourFlag(arr):
  # TO DO
  wave_array = np.array([0]*len(arr))
  wave_array[0] = arr[0]

  if arr[0] % 2 == 0:

    odd_keeper = 1
    for i in range(1, len(arr)):
      if arr[i] % 2 == 1:
        wave_array[odd_keeper] = arr[i]
        odd_keeper += 2

    even_keeper = 2

    for i in range(1, len(arr)):
      if arr[i] % 2 == 0:
        wave_array[even_keeper] = arr[i]
        even_keeper += 2

  else:

    odd_keeper = 2
    for i in range(1, len(arr)):
      if arr[i] % 2 == 1:
        wave_array[odd_keeper] = arr[i]
        odd_keeper += 2

    even_keeper = 1

    for i in range(1, len(arr)):
      if arr[i] % 2 == 0:
        wave_array[even_keeper] = arr[i]
        even_keeper += 2

  return wave_array
```

```python
#part 2

#task-1 (finding mean)

given_array = np.array([10, 8, 13, 9, 14, 25, -5, 20, 7, 7, 4])

def meanFinder(arr):
  sum = 0
  for num in arr:
    sum += num
  mean = sum/len(arr)
  return mean

function_call_for_mean = meanFinder(given_array)

print(f'The mean of the numbers is: {function_call_for_mean}')





#task 2 - find standard deviation

import math

def standardDeviationFinder(arr):

  diff = 0

  mean = meanFinder(given_array)

  for elem in arr:

    diff += ( (mean - elem)**2 )

  sd = round( math.sqrt (diff / (len(arr)-1)), 2)

  return sd

function_call_for_standard_deviation = standardDeviationFinder(given_array)
print(f'The standard deviation is: {function_call_for_standard_deviation}')






#task 3 - find specific data points with respect to standard deviation

def sd_specific_data_points_finder(arr):
  sd = standardDeviationFinder(given_array)
  mean = meanFinder(given_array)

  count = 0

  for elem in arr:

    if (elem >= mean + 1.5 * sd) or (elem <= mean - 1.5 * sd):
      count += 1

  new_array = np.array([0]*count)

  count_2 = 0

  for elem in arr:

    if (elem >= mean + 1.5 * sd) or (elem <= mean - 1.5 * sd):

      new_array[count_2] = elem
      count_2 += 1

  return new_array

final_func_call = sd_specific_data_points_finder(given_array)

print(f'New array: {final_func_call}')
```
